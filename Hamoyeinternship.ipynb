```python
qi_x  = np.array(x_train["T2"]).reshape(-1,1) #just putting this into an array to make it 2 dimensional
qi_x.ndim #RETURNS THE NO OF DIMENSIONS OF THE ARRAY
qi_x.shape
x_train["T2"].shape
```

```python
qi_y  = (x_train["T6"])
qi_y.ndim
```

```python
qi_x_train, qi_x_test, qi_y_train, qi_y_test = train_test_split(qi_x, qi_y, test_size = 0.3, random_state = 42)
```

```python
newlinear_model = LinearRegression().fit(qi_x_train, qi_y_train)
```

```python
newlinear_predict = newlinear_model.predict(qi_x_test)
newlinear_predict
```

```python
mae = mean_absolute_error(qi_y_test, newlinear_predict)
mae.round(2)
```

```python
qi_rss = np.sum(np.square(qi_y_test - newlinear_predict))
round(qi_rss, 2)
```

```python
qi_rmse = np.sqrt(mean_squared_error(qi_y_test, newlinear_predict))
round(qi_rmse, 3)
```

```python
# Coefficient of Determination is the same as r2_score
qi_cod = r2_score(newlinear_predict, (qi_y_test))
round(qi_cod, 2)
```
